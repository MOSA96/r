# Visualizacón de datos con R

En este notebook aprenderemos a visualizar datoscon ggplot2 

```r
#Importamos librería

library(tidyverse)
```

```r
#Importamos nuestro dataframe

head(mpg)
```

```r
ggplot(data=mpg)+geom_point( mapping = aes(x = displ, y = hwy))
```

Describamos más un poco en código <code>ggplot(data=mpg)+geom_point(mapping = aes(x = displ, y = hwy)) </code>

<code>ggplot(data=mpg)</code> En esta parte le indicamos que información tiene que leer.

<code>geom_point()</code> Le indicamos que es un gráfico de puntos.

<code>mapping = aes(x = displ, y = hwy)</code>  Le damos los ejes a gráficas.

Si agregamos el argumento <code>colour=class</code> podemos clasificar dependiendido de la clase.

Además con <code>options()</code> podemos especificar el tamaño de nuestra figura.

```r
options(repr.plot.width=10, repr.plot.height=4)
ggplot(data=mpg)+geom_point( mapping = aes(x = displ, y = hwy, colour=class))
```

En el gráfico anterior gráficamos displ(tamaño de motor) vs displ(eficiencia). ¿Qué pasa si queres hacer una gráfica de tamaño de motor contra el número de cilindros? 

Facíl solo especificamos <code>aes()</code> los ejes a gráfica en este caso será <code>aes(x=displ,y=cyl)</code>


```r
options(repr.width=10, repr.plot.height=3)
ggplot(data=mpg) + geom_point(mapping = aes(x=displ, y=cyl, colour=class))
```

De este gráfico podemos deducir que la mayoría de las SUV tienen 8 cilindros y los autos de tamaño tienen 6 cilindros.

Estas notas son para una visualización de datos pero podemos graficar también podemos gráficar funciones. Podemos graficar con <code>plot</code> o <code>curve</code> dependiendo del caso

```r
plot(sin, -pi, 2*pi)
```

```r
curve(x^2, from=1, to=10)
```
